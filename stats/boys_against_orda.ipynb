```python
import numpy as np
import pandas as pd
```

```python
import json
with open("transformed.json", "r") as fin:
  measure_json = json.loads(fin.read())
```

```python
measure_values = [y["length"] for x in measure_json.values() for y in x["measures"]]
```

```python
import matplotlib.pyplot as plt
import scipy.stats as st
%matplotlib inline
plt.hist(measure_values, density=True, bins=50)
plt.grid(True)
plt.xlabel("Length")

mn, mx = plt.xlim()
plt.xlim(mn, mx)
kde_xs = np.linspace(mn, mx, 300)
kde = st.gaussian_kde(measure_values)
plt.plot(kde_xs, kde.pdf(kde_xs), label="PDF")
```
